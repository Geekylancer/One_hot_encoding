#                  One - Hot Encoding with multiple categorical features ----

```
# Import Libraries

import pandas as pd
import numpy as np
```

```
# Import kaggle dataset : https://www.kaggle.com/arsenland/mercedesbenz-greener-manufacturing?select=test.csv

df = pd.read_csv('mercedes.csv',usecols = ['X0','X1','X2','X3','X4','X5'])
df.head(10)
```

```
df.columns
```

```
# Printing each column with unique number of labels

for i in df.columns:                        
    print(i,':',len(df[i].unique()),'labels')
```

```
pd.get_dummies(df,drop_first=True).shape
```

Performing one hot encoding in such a huge number of columns will increase the number of dimensions that leads to curse of dimensionality.Thereby, creates impact to our accuracy level. So, we take top 10 labels of each columns occuring most frequent number of times.

```
df.X0.value_counts().head(20)
```

```
# Taking top 10 most frequently occuring labels of column 'X0'

top10_0 = [j for j in df.X0.value_counts().head(10).index]              
top10_0
```

```
for k in top10_0:
    df[k] = np.where(df['X0'] == k,1,0)
df[['X0']+top10_0].head(30)
```

```
def ohe(data,top,var):
    for k in top:
        data[var+'_'+k] = np.where(data[var]==k,1,0)

df = pd.read_csv('mercedes.csv',usecols = ['X0','X1','X2','X3','X4','X5'])
ohe(df,top10_0,'X0')
df.head()
```

```
# Considering the remaining columns

top10_1,top10_2,top10_3,top10_4,top10_5 = [j for j in df.X1.value_counts().head(10).index],[x for x in df.X2.value_counts().head(10).index],[y for y in df.X3.value_counts().head(10).index],[z for z in df.X4.value_counts().head(10).index],[w for w in df.X5.value_counts().head(10).index]
ohe(df,top10_1,'X1')
ohe(df,top10_2,'X2')
ohe(df,top10_3,'X3')
ohe(df,top10_4,'X4')
ohe(df,top10_5,'X5')
df.head()
```
